> let us see that how we can make a program that can help us to make a workout plan for us.

```python
import pandas as pd
import numpy as np
```

+ here we have our data which has the information about:
    + Body Part.
    + Type of Muscle.
    + Workout.
    + Sets.
    + Reps per Set.

* we will use this data to make a workout plan for us.

```python
data = 'Workout.csv'

df_workout = pd.read_csv(data)

df_workout
```

It has been observed that the ideal workout plan is to workout 4 days a week and rest for 3 days. So, we will make a workout plan for 4 days a week.

* we have 51 exercises in our data so we will divide them into 4 days a week according to the body part such that it is parted by push, pull, push, pull movement.
* Let us see how we can do this.

```python
legs_shoulders = df_workout[(df_workout['Body Part'] == 'Legs') | (df_workout['Body Part'] == 'Shoulders')]
legs_shoulders
```

We have Divided the body parts into 4 parts

* Push: Chest, Triceps, Shoulders
* Pull: Back, Biceps, Forearms
* Legs: Legs
* Abs: Abs

* We have combined the body parts into 4 parts and we will make a workout plan for 4 days a week.

The above table shows the workout plan of Legs and Shoulders for the first day.

```python
back_biceps = df_workout[(df_workout['Body Part'] == 'Back') | (df_workout['Type of Muscle'] == 'Biceps')]
back_biceps
```

We have Divided the body parts into 4 parts

* Push: Chest, Triceps, Shoulders

* Legs: Legs

```python
print(f'exercise count\n1.) Chest-Triceps:\n{chest_triceps.iloc[:,:2].count()}.\n2.) Back-Biceps:\n{back_biceps.iloc[:,:2].count()}.\n3.) Legs-Shoulders:\n{legs_shoulders.iloc[:,:2].count()}.\n4.) Abs-Forearms:\n{abs_forearms.iloc[:,:2].count()}.')
```

Above is the count of the exercises that we have in our data for each body part.

```python
value_chest_tricep = chest_triceps.iloc[:,:2].value_counts()
```

```python
value_back_bicep = back_biceps.iloc[:,:2].value_counts()
```

```python
value_legs_shoulders = legs_shoulders.iloc[:,:2].value_counts()
```

```python
value_abs_forearms = abs_forearms.iloc[:,:2].value_counts()
```

```python
import matplotlib.pyplot as plt

x = ['value_chest_tricep', 'value_back_bicep', 'value_legs_shoulders', 'value_abs_forearms']
y = [np.sum(value_chest_tricep.tolist()), np.sum(value_back_bicep.tolist()), np.sum(value_legs_shoulders.tolist()), np.sum(value_abs_forearms.tolist())]
# y = [value_chest_tricep.tolist(), value_back_bicep.tolist(), value_legs_shoulders.tolist(), value_abs_forearms.tolist()]
plt.figure(figsize=(15, 10))
plt.bar(x,y)
```

You can see that there is an ideal distribution of the exercises for each body part so that it can grow more.

```python
per_body_part = df_workout.groupby(['Body Part', 'Sets'])
per_body_part.first()
```

Here we have an high level view on our schedule and reps per set grouped to each body part.

```python
df_chest_workouts = df_workout[df_workout['Body Part'] == 'Chest']

df_back_workouts = df_workout[df_workout['Body Part'] == 'Back']

df_legs_workouts = df_workout[df_workout['Body Part'] == 'Legs']

df_shoulders_workouts = df_workout[df_workout['Body Part'] == 'Shoulders']

df_abs_workouts = df_workout[df_workout['Body Part'] == 'Abs']

df_forearms_workouts = df_workout[df_workout['Body Part'] == 'Forearms']

df_biceps_workouts = df_workout[df_workout['Type of Muscle'] == 'Biceps']

df_triceps_workouts = df_workout[df_workout['Type of Muscle'] == 'Triceps']
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.pie(value_chest_tricep.tolist(), labels=value_chest_tricep.index.tolist(), autopct='%1.1f%%')
plt.title('Chest-Triceps')

plt.subplot(2, 2, 2)
plt.pie(value_back_bicep.tolist(), labels=value_back_bicep.index.tolist(), autopct='%1.1f%%')
plt.title('Back-Biceps')

plt.subplot(2, 2, 3)
plt.pie(value_legs_shoulders.tolist(), labels=value_legs_shoulders.index.tolist(), autopct='%1.1f%%')
plt.title('Legs-Shoulders')

plt.subplot(2, 2, 4)
plt.pie(value_abs_forearms.tolist(), labels=value_abs_forearms.index.tolist(), autopct='%1.1f%%')
plt.title('Abs-Forearms')


plt.show()
```
